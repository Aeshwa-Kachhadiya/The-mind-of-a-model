```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/Aeshwa-Kachhadiya/The-mind-of-a-model/main/dataset/linkedin_post_performance.xlsx'

df = pd.read_excel(url)
print('✅ Dataset loaded!')
```

```python
# Shape of the dataset
print('SHAPE:')
df.shape
```

```python
# First 5 rows
print('FIRST LOOK:')
df.head()
```

```python
# Data types
print('DATA TYPES:')
df.dtypes
```

```python
# Check missing values
print('MISSING VALUES:')
df.isnull().sum()
```

```python
# Statistical summary
print('STATISTICAL SUMMARY:')
df.describe()
```

```python
# Target variable distribution
print('TARGET VARIABLE DISTRIBUTION:')
df['went_viral'].value_counts()
```

```python
# Handle missing values
# Fill numerical columns with median
numerical_cols = ['word_count', 'account_followers',
                  'likes_first_hour', 'impressions_first_hour']

for col in numerical_cols:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)

print('AFTER FILLING MISSING VALUES:')
print(df.isnull().sum())
```

```python
# Check and remove duplicates
print(f'Duplicate rows: {df.duplicated().sum()}')

df.drop_duplicates(inplace=True)

print(f'✅ Shape after removing duplicates: {df.shape}')
```

```python
# Final dataset check
print('FINAL DATASET:')
print(f'Shape: {df.shape}')
print(f'Missing values: {df.isnull().sum().sum()}')
df.head()
```
